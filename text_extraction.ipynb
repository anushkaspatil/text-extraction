```python
def extract_img_txt(img_path):
    try:
        text = pytesseract.image_to_string(img_path)
        return text
    except Exception as e:
      print(e)
      return "error"

def extract_text_from_pdf(pdf_path):
    images = convert_from_path(pdf_path)
    text = ""

    for i, image in enumerate(images):
        gray = cv2.cvtColor(np.array(image), cv2.COLOR_BGR2GRAY)
        _, thresh = cv2.threshold(gray, 150, 255, cv2.THRESH_BINARY)
        page_text = pytesseract.image_to_string(thresh)
        text += f"Page {i+1}:\n{page_text}\n\n"

    return text
```

```python
image_path = '/content/drive/MyDrive/assignment internship/img3.jpg'

# Extract text from image
image_text = extract_img_txt(image_path)
print("Extracted text:")
# print(image_text)

pdf_path = '/content/drive/MyDrive/Anushka_Patil_Resume1.pdf'

# Extract text from PDF
pdf_text = extract_text_from_pdf(pdf_path)
print("Text extracted from PDF:")
print(pdf_text)
```
